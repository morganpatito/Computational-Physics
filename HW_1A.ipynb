### Introduction to Vectors and NumPy

Let's first see how to create vectors in Python using the NumPy library. We store the vector components in a NumPy array.  Run the following code cell to execute it by pressing Shift-Enter or clicking the play button.

```python
import numpy as np

# Define 3D vectors using NumPy arrays
A = np.array([1, 2, 3])
B = np.array([-1, 0, 2])

# Display the vectors
print("Vector A = ", A)
print("Vector B = ", B)
```

## Magnitude of a Vector

The magnitude of a vector, also known as its Euclidean norm or "length," is calculated using the Pythagorean theorem. For a 3D vector $ \mathbf{A} = [A_x, A_y, A_z] $, its magnitude is given by:

$$ A=|\mathbf{A}| = \sqrt{A_x^2 + A_y^2 + A_z^2} .$$

In NumPy, you can use NumPy's Linear Algebra module and the `norm()` function to calculate a vector's magnitude. The command looks like this:  `np.linalg.norm()` .

```python
import numpy as np

# Calculate the magnitude of A and B
A_mag = np.linalg.norm(A)
B_mag = np.linalg.norm(B)

# Display the results
print(f"Magnitude of A = {A_mag:.2f}")
print(f"Magnitude of B = {B_mag:.2f}")
```

## Dot Product

The dot product (also known as the scalar product) takes two vectors and returns a single scalar value. It's often used to find the angle between two vectors or to project one vector onto another.

For two vectors $ \mathbf{A} = [A_1, A_2, A_n] $ and $ \mathbf{B} = [B_1, B_2, .B_3] $, the dot product is defined as:

$$ \mathbf{A} \cdot \mathbf{B} = \sum_{i=1}^{n} A_i B_i = A_1 B_1 + A_2 B_2 +A_3 B_3$$

In NumPy, you can use `np.dot(A,B)` or the ` @` operator to write `A @ B`:

```python
import numpy as np

# Finding the dot product between A and B
prod = np.dot(A,B)
print(f"Dot product A and B = {prod}")

# This example shows how to use the @ operator for dot products
prod2 = A @ B
print(f"We get the same result using '@':  A @ B = {prod2}")
```

## Angle Between Two Vectors

The dot product can be written in two ways:
$$ \mathbf{A} \cdot \mathbf{B} = \sum_{i=1}^{n} A_i B_i = A_1 B_1 + A_2 B_2 +A_3 B_3$$
and
$$ \mathbf{A} \cdot \mathbf{B}  = A B \cos \theta. $$

The Python command `np.dot()` uses the first method to calculate the dot product. We can then use the second definiton to solve for the angle $\theta$ once the dot product is known:
$$\theta = \cos^{-1} \left( \frac{ \mathbf{A} \cdot \mathbf{B}}{AB} \right) $$


```python
import numpy as np

# Calculate the magnitude of v1 and v2
A_mag = np.linalg.norm(A)
B_mag = np.linalg.norm(B)

# Calculate the angle of v1 and v2
theta = np.acos(np.dot(A,B) / (A_mag * B_mag))

# Convert to degrees
theta_deg = np.rad2deg(theta)

# Display the results
print(f"Angle between v1 and v2 = {theta_deg:.2f} degrees")
```

## Create a Unit Vector $\widehat{\mathbf{a}}$ from a Vector $\mathbf{A}$

To create a unit vector pointing in the same direction as a vector $\mathbf{A}$, we simply divide the vector by its magnitude to force it to have unit length:
$$ \widehat{\mathbf{a}}  = \frac{\mathbf{A}}{A},$$
where we use the "hat" ($\;\widehat{}\;$) to designate a unit vector.



```python
# create a unit vector that points in the direction of vector A
a = A / A_mag
print(f"If A = {A}, the corresponding unit vector is {a}")
```

### Cross Product

The cross product (also known as the vector product) takes two 3D vectors and returns a new 3D vector that is perpendicular to both original vectors. The magnitude of the cross product is related to the area of the parallelogram formed by the two vectors.

For two 3D vectors $ \mathbf{A} = [A_x, A_y, A_z] $ and $ \mathbf{B} = [B_x, B_y, B_z] $, the cross product is defined as:

$ \mathbf{A} \times \mathbf{B} = [ (A_y B_z - A_z B_y), -(A_x B_z - A_z B_x ), (A_x B_y - A_y B_x) ] $

In NumPy, you can use `np.cross(A,B)`.

```python
import numpy as np

# Calculate the cross product C = A x B
C = np.cross(A,B)

print(f"Cross product A x B = {C}")
```

## ✅ Skill Check 4 (15 points)

Verify the following vector identities for the following vectors  $\mathbf{A} = [1,3,2]$, $\mathbf{B} = [-2,1,2]$, $\mathbf{A} = [3,-2,1]$:
* $\mathbf{A} \times \mathbf{B} = -\mathbf{B} \times \mathbf{A}$
* $\mathbf{A} \times (\mathbf{B} + \mathbf{C} ) = \mathbf{A} \times \mathbf{B} + \mathbf{A} \times \mathbf{C}$
* $\mathbf{A} \cdot (\mathbf{B}\times \mathbf{C}) = (\mathbf{A} \times \mathbf{B})\cdot \mathbf{C}$
* $\mathbf{A} \times (\mathbf{B}\times \mathbf{C}) = (\mathbf{A} \cdot \mathbf{C})\mathbf{B} - (\mathbf{A}\cdot \mathbf{B}) \mathbf{C}$
* $\mathbf{A} \cdot (\mathbf{A} \times \mathbf{B})=0$
* $\mathbf{A} \times \mathbf{A} = 0$

```python
# your code here
import numpy as np

A = np.array([ 1,  3, 2], dtype=float)
B = np.array([-2,  1, 2], dtype=float)
C = np.array([ 3, -2, 1], dtype=float)

def show(name, ok):
    print(f"{name}: {'✅ True' if ok else '❌ False'}")

# 1) A×B = −B×A
lhs1 = np.cross(A, B)
rhs1 = -np.cross(B, A)
show("1) A×B = −B×A", np.allclose(lhs1, rhs1))

# 2) A×(B+C) = A×B + A×C
lhs2 = np.cross(A, B + C)
rhs2 = np.cross(A, B) + np.cross(A, C)
show("2) A×(B+C) = A×B + A×C", np.allclose(lhs2, rhs2))

# 3) A⋅(B×C) = (A×B)⋅C
lhs3 = np.dot(A, np.cross(B, C))
rhs3 = np.dot(np.cross(A, B), C)
show("3) A⋅(B×C) = (A×B)⋅C", np.isclose(lhs3, rhs3))

# 4) A×(B×C) = (A⋅C)B − (A⋅B)C
lhs4 = np.cross(A, np.cross(B, C))
rhs4 = (np.dot(A, C))*B - (np.dot(A, B))*C
show("4) A×(B×C) = (A⋅C)B − (A⋅B)C", np.allclose(lhs4, rhs4))

# 5) A⋅(A×B) = 0
lhs5 = np.dot(A, np.cross(A, B))
show("5) A⋅(A×B) = 0", np.isclose(lhs5, 0))

# 6) A×A = 0
lhs6 = np.cross(A, A)
show("6) A×A = 0", np.allclose(lhs6, np.zeros(3)))

print("\nExtra values (optional):")
print("A×B =", np.cross(A,B))
print("B×A =", np.cross(B,A))
print("A⋅(B×C) =", lhs3)
print("(A×B)⋅C =", rhs3)
print("A×(B×C) =", lhs4)
print("RHS triple-product identity =", rhs4)
```

## ✅ Skill Check 5 (10 points)

Consider a point mass at Cartesian coordinates $\mathbf{r} = [2, 1]$m with velocity $\mathbf{v} = [-3, 3]$m/s.
* (a) Find the polar coordinates $(r,\theta)$ of this point.  
* (b) Calculate the unit vectors $\widehat{\mathbf{r}}$ and $\widehat{\mathbf{\theta}}$ given $(r,\theta)$.
* (c) Calcuate the polar velocity components $(v_r,v_\theta)$.



```python
# your code here
import numpy as np

r_vec = np.array([2.0, 1.0])     # meters
v_vec = np.array([-3.0, 3.0])    # m/s

# (a) polar coordinates (r, theta)
r = np.linalg.norm(r_vec)
theta = np.arctan2(r_vec[1], r_vec[0])   # radians
theta_deg = np.rad2deg(theta)

# (b) unit vectors rhat and thetahat
rhat = np.array([np.cos(theta), np.sin(theta)])
thetahat = np.array([-np.sin(theta), np.cos(theta)])

# (c) polar velocity components (vr, vtheta)
vr = np.dot(v_vec, rhat)
vtheta = np.dot(v_vec, thetahat)

print("(a) r =", r, "m")
print("    theta =", theta, "rad  =", theta_deg, "deg")
print("(b) rhat =", rhat)
print("    thetahat =", thetahat)
print("(c) vr =", vr, "m/s")
print("    vtheta =", vtheta, "m/s")
```
